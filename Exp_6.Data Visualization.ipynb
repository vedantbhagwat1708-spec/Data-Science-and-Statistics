```python
#import
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

# Line Chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()

```

# Bar Chart

```python
plt.bar(x,y,color='yellow',edgecolor='black')
plt.title(" Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Scatter Plot

```python
a=(1,5,8,6,3,7,9,4)
b=(20,12,23,24,32,45,43,56)


plt.scatter(a,b,color='yellow',edgecolor='black')
plt.title(" scatter plot")
plt.xlabel("a axis")
plt.ylabel("b axis")
plt.show()
```

# Histogram

```python
H=(1,2,2,3,3,3,4,2,5,5,5,5,6,77,76,65,65,88,88,90)
```

```python
plt.hist(H)
plt.show()
```
